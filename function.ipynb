6. Explain the concept of generators in Python and how they are defined.
  - A generator in python is a special type of iterator that generates values on the fly instead of storing of them in memory.it is defined using a function with the yield keyword.
  - A generator is defined like a normal function but uses yield instead of return.

```python
def count_up_to(n):
  count = 1
  while count <= n:
    yield count
    count +=1
# using the generator
gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
   1. map(funcion, iterable)
   - applies a function to each element in an iterable.
   - Returns: A new iterable with modified values.
   - use case: Transform elements (e.g.,squaring numbers,converting strings to uppercase).
   2. filter(function,iterable)
   - select elements that satisfy a condition(i.e., function return True)
   - Return: A filtered iterable with only the element that meet the condition.
   - use case: Extract specific values(e.g., keeping only even numbers).
   3. reduce(function, iterable)(from functools module)
   - Reduces all elements into a single value by applying a function cumulatively.
   - Returns: A sigle computed value(e.g., sum,product,maximum).
   - use case:compute cumulative result(e.g., sum of number,finding the product).

```python
#6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
  for i in range(exponent+1):
    yield 2 ** i

for num in power_of_two(5):
  print(num, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
  with open(filename, 'r') as file:
    for line in file:
      yield line
```

```python
#8.  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple
data = [("alice", 25),("bob", 20),("david", 18),("charlie", 23)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```
